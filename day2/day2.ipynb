```python
max_fre = [int(re.findall(r'-(\d*)',i)[0]) for i in fre]
max_fre
```

```python
word = [re.findall(r'(\w):',i)[0] for i in input_list]
```

```python
password = [re.findall(r'(?<=\: ).+',i)[0] for i in input_list]
password
```

```python
df = pd.DataFrame()
df['min_fre'] = min_fre
df['max_fre'] = max_fre
df['word'] = word
df['password'] = password
df
```

```python
df['label'] = ''
for i in range(len(df)):
    char_count = 0
    for char in df['password'][i]:
        if df['password'][i][df['min_fre'][i]-1] == df['word'][i] and df['password'][i][df['max_fre'][i]-1] == df['word'][i]:
            df['label'][i] = 0
        elif df['password'][i][df['min_fre'][i]-1] != df['word'][i] and df['password'][i][df['max_fre'][i]-1] == df['word'][i]:
            df['label'][i] = 1
        elif df['password'][i][df['min_fre'][i]-1] == df['word'][i] and df['password'][i][df['max_fre'][i]-1] != df['word'][i]:
            df['label'][i] = 1
        else:
            df['label'][i] = 0
```
